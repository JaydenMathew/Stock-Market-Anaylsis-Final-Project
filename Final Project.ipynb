#### Question 1: From the years of 2008 to 2023, which company (AAPL, AMZN, TSLA, GOOG) had the highest percent change overall?

```python
# Group by 'COMPANY' and calculate the percentage change in adjusted closing price
df_filtered = df
percentage_change = df_filtered.groupby('COMPANY')['ADJ CLOSE'].apply(lambda x: ((x.iloc[-1] - x.iloc[0]) / x.iloc[0]) * 100)

# Create a table with the percent changes
percent_changes_table = pd.DataFrame({
    'Company': ['APPLE', 'AMAZON', 'TESLA', 'META'],
    'Percent Change': [percentage_change.get('APPLE', 0),
                       percentage_change.get('AMAZON', 0),
                       percentage_change.get('TESLA', 0),
                       percentage_change.get('META', 0)]
})
```

```python
# Display the table
print("Table of Percent Changes:")
print("")
print(percent_changes_table.to_string(index=False))
print("")

# Find the value of the company with the highest percent change
max_percent_change = max(percentage_change)

# Determine which company had the highest percent change
highest_performer = percentage_change.idxmax()

print(highest_performer, "had the HIGHEST percent change of:", max_percent_change,"% overall from 2008 to 2023.")
```

```python
# Plot the table/percentage change
plt.figure(figsize=(10, 6))
plt.bar(percent_changes_table['Company'], percent_changes_table['Percent Change'], color=['blue', 'green', 'red', 'orange'])
plt.title('Percentage Change in Adjusted Closing Price (2008-2023)')
plt.xlabel('Company')
plt.ylabel('Percentage Change (%)')
plt.yticks(range(0, int(max_percent_change) + 100, 100))
plt.grid(axis='y')
plt.show()
```

### It is clear from the graph and table above that Tesla had the highest percent change from 2018-2023 (of about 911%) and Amazon had the lowest percent change (of about 94%). It is also clear that out of the middle two companies, Apple had a higher percent change than Meta, both of which are about 346% and 152% respectively.

#### Question 2: What was the average stock price since 2018 for all of the companies?

```python
# Group by 'COMPANY' and calculate the average adjusted closing price since 2018 for each company
average_price_since_2018 = df_filtered.groupby('COMPANY')['ADJ CLOSE'].mean()

# Create a table with the average prices since 2018 for each company
average_prices_table = pd.DataFrame({'Company': average_price_since_2018.index, 'Average Price Since 2018': average_price_since_2018.values})

# Display the table without index
print("Average Stock Prices Since 2018:")
print("")
print(average_prices_table.to_string(index=False))
```

```python
# Define colors for each company
company_colors = {'APPLE': 'tab:blue', 'AMAZON': 'tab:orange', 'TESLA': 'tab:green', 'META': 'tab:red'}

# Plot the trend of stock prices over time for each company
plt.figure(figsize=(12, 6))
for company, data in df_filtered.groupby('COMPANY'):
    plt.plot(data['DATE'], data['ADJ CLOSE'], label=company, color=company_colors[company])

# Plot the average line for each company with matching colors
for company in average_price_since_2018.index:
    average = average_price_since_2018[company]
    plt.axhline(y=average, color=company_colors[company], linestyle='--', label=f'Avg. ({company})')

# Add labels and legend
plt.title('Stock Price Trend Since 2018')
plt.xlabel('Date')
plt.ylabel('Adjusted Closing Price')
plt.legend()
plt.grid(True)
plt.show()
```

### From the table and graph above, it is clear that the companies have had many peaks and valleys between 2018 to 2023. In terms of average stock value, Meta has had the highest average stock price (at around 231 dollars per share) and Apple has had the lowest average stock price (at around 118 per share). #### Question 5: Which company had the highest change in stock value held in a singular day? (List this from highest change to lowest change)


```python
# Calculate the absolute difference between opening and closing prices for each day for each company
df['PRICE_CHANGE'] = abs(df['CLOSE'] - df['OPEN'])

# Find the maximum price change for each company
max_price_change = df.groupby('COMPANY')['PRICE_CHANGE'].max()

# Sort the companies based on their maximum price change in descending order
sorted_companies = max_price_change.sort_values(ascending=False)
```

```python
# Plot max price change values
plt.figure(figsize=(10, 6))
sorted_companies.plot(kind='bar', color='lightseagreen')
plt.title('Highest Change in Stock Value Held in a Single Day by Company')
plt.xlabel('Company')
plt.ylabel('Price Change')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
